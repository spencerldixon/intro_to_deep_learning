```python
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

dataframe = pd.read_fwf('brain_body.txt')

x_values = dataframe[['Brain']]
y_values = dataframe[['Body']]

dataframe
```

```python
model = linear_model.LinearRegression()
model.fit(x_values, y_values)
```

```python
plt.scatter(x_values, y_values)
plt.plot(x_values, model.predict(x_values))
plt.show()
```

# Challenge Dataset

```python
dataframe = pd.read_csv('challenge_dataset.txt', header=None) # Don't use first row as headers
dataframe.columns = ["X", "Y"]

x_values = dataframe[["X"]]
y_values = dataframe[["Y"]]

dataframe
```

```python
# Plot this to a graph to begin with

plt.scatter(x_values, y_values)
plt.show()
```

```python
# Make a model

model = linear_model.LinearRegression()
model.fit(x_values, y_values)

y_values_predicted = model.predict(x_values)

plt.scatter(x_values, y_values)
plt.plot(x_values, y_values_predicted)
plt.show()
```

```python
print("Mean squared error: %.2f #" % mean_squared_error(x_values, y_values_predicted))
```
